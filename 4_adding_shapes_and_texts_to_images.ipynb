# Opening images with OpenCV 

#### (that section was quite troublesome and made me go nuts trying to run it on the Learning_OpenCV notebook so i decided to give it its own file

```python
import cv2 as cv
```

```python
puppy_img = cv.imread("puppy.jpg")

puppy_img = cv.resize(puppy_img,(1280,720))
while True :
    
    cv.imshow("Puppy",puppy_img)
    
    if cv.waitKey(1) & 0xff == 27: 
        break
        
    ''' the condition is not that complicated,
    it just say that if we waited more the 1 ms and pressed the esc button whose code is 27 then break the loop ( thus close the image),
    that way the kernal wouldn't crash and we won't force quit everytime we open some image ''' 
    
cv.destroyAllWindows()
```

# Drawing shapes & adding text on an image

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
```

```python
black_img = np.zeros(shape = (720,1280,3), dtype = np.int16)
black_img.shape
```

```python
plt.imshow(black_img)
```

## Drawing a rectangle

```python
rr_black_img = black_img.copy()
rr_black_img = cv.rectangle(rr_black_img, pt1 = (300,100), pt2 = (900,600), color = (255,0,0), thickness = 10) 
```

#### i named it with a different name so that i don't get confused with the content of the image, but the rectangle function indeed modify the actual given image

```python
plt.imshow(rr_black_img)
```

#### From now on i will be using different names for each shape i drow, copy the original black image to it and then apply the drawn shape

## Drawing a Circle

```python
pc_black_img = black_img.copy()
pc_black_img = cv.circle(pc_black_img, center = (640,360), radius = 100, color = (255,0,255), thickness = 7 )
```

```python
plt.imshow(pc_black_img)
```

## Drawing a filled circle (thickness = -1)

```python
rc_filled_black_img = black_img.copy()
rc_filled_black_img = cv.circle(rc_filled_black_img, center = (400,360), radius = 100, color = (255,0,0), thickness = -1 )
```

```python
plt.imshow(rc_filled_black_img)
```

## Drawing a Line

```python
gl_black_img = black_img.copy()
gl_black_img = cv.line(gl_black_img, pt1 = (0,720), pt2 = (1280,0), color = (0,255,0), thickness = 5)
```

```python
plt.imshow(gl_black_img)
```

### You must be wondering now ! , why i always copy the black image before i modify the image by any shape 
### i will give the answer to you shortly in the so called black_img of ours

```python
cv.rectangle(black_img, pt1 = (300,100), pt2 = (900,600), color = (255,0,255), thickness = 10) 
cv.circle(black_img, center = (200,600), radius = 100, color = (122,200,150), thickness = -1 )
cv.circle(black_img, center = (600,360), radius = 100, color = (200,55,100), thickness = 7 )
cv.circle(black_img, center = (1080,120), radius = 100, color = (75,100,170), thickness = -1 )
cv.line(black_img, pt1 = (0,720), pt2 = (1280,0), color = (0,255,0), thickness = 5)
```

```python
plt.imshow(black_img)
```

### You can see it now, i didn't assign any new image variable to the changes i've made, the changes is don't to my inserted image permenantly

## Writing texts on images

```python
font = cv.FONT_ITALIC
cv.putText(black_img, "this is kinda small space in here" , org = (320,710) , fontFace = font, fontScale = 2.3 , color = (255,255,0) , thickness = 5,
           lineType = cv.LINE_AA)
plt.imshow(black_img)
```

## Drawing custom shapes (polylines)

```python
vertices = np.array([ [250,200], [750,800], [1500,600], [1250,400] ], dtype = np.int32)
```

```python
pts.shape
```

```python
pts
```
