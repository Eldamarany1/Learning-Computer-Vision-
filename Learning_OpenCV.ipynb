# OpenCV library and dealing with images using it

## importing librarys

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2 as cv
```

## reading images with opencv

```python
puppy_img = cv.imread("puppy.jpg")
```

```python
type(puppy_img)
```

```python
puppy_img
```

```python
puppy_img.shape
```

```python
plt.imshow(puppy_img)
```

#### Note : the reason that the image of puppy is like this is that the matplotlib library uses RGB color format but in the other hand OpenCV library uses the BGR color format (RGB but reversed) so when we call an image opened with OpenCV with matplotlib then that results

## How to fix the image (converting BRG to RGB)

```python
puppy_img = cv.cvtColor(puppy_img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(puppy_img)
```

## Making the image Grayscale

```python
gray_puppy_img = cv.imread("puppy.jpg",cv.IMREAD_GRAYSCALE)
```

```python
plt.imshow(gray_puppy_img,cmap='gray')
```

## Resizing the image using OpenCV

```python
resized_puppy_img = cv.resize(puppy_img,(1280,720))
```

```python
resized_puppy_img.shape
```

```python
plt.imshow(resized_puppy_img)
```

## resizing the image using OpenCV but with ratios

```python
width_ratio = 0.5
height_ratio = 0.5
half_shaped_puppy_img = cv.resize(puppy_img,(0,0),puppy_img,width_ratio,height_ratio)
plt.imshow(half_shaped_puppy_img)
```

## Flipping an image

```python
fliped_puppy = cv.flip(puppy_img,0) # 0 for vertical flip , 1 for horizontal flip , -1 for both
plt.imshow(fliped_puppy)
```

## Saving modified images

```python
cv.imwrite("filped_puppy.png",fliped_puppy)
```

## plotting images using matplotlib with figure function and making subplots

```python
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
ax.imshow(puppy_img)
```
