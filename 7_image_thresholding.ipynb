# Image thresholding

### image thresholding is turning a colorful 3 channel color image to a grayscale (black and white) 1 channel color one

### it has so many merits as we will see

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rainbow = cv.imread("Data/rainbow.avif",0) # read the image as a gray scale image
```

```python
plt.imshow(rainbow, cmap="gray")
```

```python
thr_val, thr_rainbow = cv.threshold(rainbow, 70, 255, cv.THRESH_TRUNC)
plt.imshow(thr_rainbow, cmap = "gray")
```
