# Image thresholding

### image thresholding is turning a colorful 3 channel color image to a grayscale (black and white) 1 channel color one

### it has so many merits as we will see

```python
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rainbow = cv.imread("Data/rainbow.jpg",0) # read the image as a gray scale image
rainbow = cv.cvtColor(rainbow,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(rainbow, cmap="gray")
```

## THRESH_BINARY
### is used to limit the colors of the images into two colors (black and white)
### black if the color channel pixel value was less than the threshold value and white otherwise

```python
thr_val, thr_rainbow = cv.threshold(rainbow, 127, 255, cv.THRESH_BINARY)
plt.imshow(thr_rainbow, cmap="gray")
```

## THRESH_BINARY_INV
### The same as THRESH_BINARY but inversed

```python
thr_val, thr_rainbow = cv.threshold(rainbow, 127, 255, cv.THRESH_BINARY_INV)
plt.imshow(thr_rainbow, cmap="gray")
```

## THRESH_TRUNK
### That one is quite complicated but the idea is simple
### if the color pixel's value is more than the threshold then it will be forced back to the threshold value
### otherwise the color will not change

```python
thr_val, thr_rainbow = cv.threshold(rainbow, 127, 255, cv.THRESH_TRUNC)
plt.imshow(thr_rainbow, cmap="gray")
```

## THRESH_TOZERO
### If the color value more than the threshold it remains the same otherwise it returns black

```python
thr_val, thr_rainbow = cv.threshold(rainbow, 127, 255, cv.THRESH_TOZERO)
plt.imshow(thr_rainbow, cmap="gray")
```

## THRESH_TOZERO_INV
### If the color value more than the threshold it it returns black otherwise remains the same

```python
thr_val, thr_rainbow = cv.threshold(rainbow, 127, 255, cv.THRESH_TOZERO_INV)
plt.imshow(thr_rainbow, cmap="gray")
```
